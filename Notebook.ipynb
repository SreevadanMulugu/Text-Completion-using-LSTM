```python
import numpy as np
from keras.models import Sequential
from keras.layers import LSTM, Dense
from keras.utils import to_categorical

# Example text for simplicity; use a longer text in practice
text = ("""Once upon a time, in a land shrouded in mystery and magic, there existed a realm where valor and chivalry reigned supreme. This is the story of Arthur, a brave knight whose legend has transcended generations, inspiring countless tales of heroism and adventure. Arthur was born in a quaint village nestled at the edge of an expansive forest, a place where myths and reality intertwined. From a young age, he exhibited extraordinary courage and an unyielding sense of justice. His parents, humble farmers, recognized his unique qualities and nurtured his burgeoning talents. As he grew, Arthur's fame spread beyond the village, reaching the ears of the kingdom's rulers.

The kingdom itself was a place of both beauty and peril, ruled by a wise and benevolent king who valued strength and integrity. When Arthur was but a young man, he was summoned to the royal court. There, he proved his mettle in a series of challenges, showcasing not only his physical prowess but also his sharp intellect and unwavering moral compass. Impressed by his capabilities, the king knighted him, bestowing upon him the title of Sir Arthur.

As a knight, Arthur embarked on numerous quests that tested his limits and solidified his reputation. One of his earliest adventures involved rescuing a village plagued by a ferocious dragon. Armed with only his sword and a heart full of courage, Arthur faced the beast in a battle that would become legendary. The dragon, with scales as hard as steel and fiery breath, was a formidable opponent. Yet, Arthur's determination and bravery saw him through, and he emerged victorious, earning the eternal gratitude of the villagers.

Arthur's most notable quest, however, was his pursuit of the elusive Holy Grail. This sacred artifact, said to possess immense power and wisdom, had been lost for centuries. Driven by a deep sense of duty and a desire to bring prosperity to his kingdom, Arthur embarked on this perilous journey. Along the way, he encountered allies who became lifelong friends, as well as adversaries who tested his resolve. Through it all, his unwavering faith and noble spirit guided him, even in the face of insurmountable odds.

The kingdom, meanwhile, was not without its troubles. A shadowy sorcerer, intent on usurping the throne, cast a dark pall over the land. The sorcerer's magic was potent, and his ambitions threatened the very fabric of the realm. Recognizing the gravity of the threat, Arthur confronted the sorcerer in a battle of wits and strength. The clash was epic, with spells and swords clashing under the stormy skies. In the end, Arthur's purity of heart and indomitable will triumphed over the sorcerer's dark ambitions, restoring peace to the kingdom.

Arthur's legacy extended beyond his lifetime. His tales of bravery and chivalry were passed down through the ages, becoming the bedrock of the kingdom's culture and ethos. His life was a testament to the power of courage, integrity, and selflessness. The legend of Arthur, the noble knight, serves as an enduring reminder that even in the face of overwhelming darkness, the light of virtue can prevail.

Generations later, children would gather around the hearth to hear the stories of Sir Arthur, dreaming of their own adventures and the possibility of greatness. His story, etched in the annals of history, continues to inspire, a beacon of hope and heroism for all who hear it. Arthur's journey, marked by trials and triumphs, is a timeless tale of courage that will never fade from the hearts of those who cherish the noble ideals he embodied.""")

# The rest of your code to process this text remains the same.

text = text.lower()  # Ensure all text is in lower case

# Create a mapping of characters to integers and vice versa
chars = sorted(set(text))
char_to_num = {ch: i+1 for i, ch in enumerate(chars)}
num_to_char = {i+1: ch for i, ch in enumerate(chars)}

# Convert text to sequence of numbers
sequence = [char_to_num[ch] for ch in text if ch in char_to_num]

# Prepare data for LSTM
seq_length = 10  # Choose a sequence length
X = []
y = []

for i in range(len(sequence) - seq_length):
    X.append(sequence[i:i + seq_length])
    y.append(sequence[i + seq_length])

# Convert lists to numpy arrays
X = np.array(X)
y = np.array(y)

# One-hot encode the output variable
y = to_categorical(y, num_classes=len(char_to_num) + 1)

# Reshape X to be [samples, time steps, features]
X = np.reshape(X, (X.shape[0], X.shape[1], 1))
X = X / float(len(char_to_num) + 1)  # Normalize to the range [0, 1]

# Define LSTM model
model = Sequential()
model.add(LSTM(100, input_shape=(X.shape[1], X.shape[2])))
model.add(Dense(len(char_to_num) + 1, activation='softmax'))
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# Train the model
model.fit(X, y, epochs=250, batch_size=32)

# Predict the next sequence
def predict_next_chars(model, start_seq, num_to_char, num_predictions=100):
    start_seq = [char_to_num[ch] for ch in start_seq if ch in char_to_num]
    predicted_seq = start_seq[:]
    for _ in range(num_predictions):
        x = np.reshape(predicted_seq[-seq_length:], (1, seq_length, 1))
        x = x / float(len(char_to_num) + 1)
        prediction = model.predict(x, verbose=0)
        index = np.argmax(prediction)
        predicted_seq.append(index)
    return ''.join([num_to_char[val] for val in predicted_seq[-num_predictions:]])

start_seq = "once upon "
predicted_text = predict_next_chars(model, start_seq, num_to_char)
print(predicted_text)
```
